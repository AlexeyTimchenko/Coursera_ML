```python
import numpy as np
import scipy
from scipy import linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

grid = np.arange(1, 15.1, 0.25)
function_values = f(grid)

plt.plot(grid, function_values)
plt.show()
```

```python
def create_matrix_A(X, n):
    """
    X - массив точек, его длина - n + 1
    n - степень многочлена
    """
    A = np.zeros((n+1, n+1))
    for i in range(n+1):
        for j in range(n + 1):
            A[i, j] = X[i]**j
    return A

def create_column_b(X, n):
    b1 = np.array(X)
    return f(b1)

def polynome(point, weights):
    result = 0
    for i in range(len(weights)):
        result += weights[i] * (point**i)
    return result
        
```

```python
# 1-ая степень x1 = 1, x2 = 15

b1 = create_column_b([1, 15], 1)
A1 = create_matrix_A([1, 15], 1)

weights_1 = linalg.solve(A1, b1)

y1 = polynome(grid, weights_1)

plt.plot(grid, y1)
plt.show()
```

```python
X2 = [1, 8, 15]

b2 = create_column_b(X2, 2)
A2 = create_matrix_A(X2, 2)

weights_2 = linalg.solve(A2, b2)

y2 = polynome(grid, weights_2)

plt.plot(grid, y2)
plt.plot(grid, function_values)
plt.show()
```

```python
X3 = [1, 4, 10, 15]

b3 = create_column_b(X3, 3)
A3 = create_matrix_A(X3, 3)

weights_3 = linalg.solve(A3, b3)

y3 = polynome(grid, weights_3)

plt.plot(grid, y3)
plt.plot(grid, function_values)
plt.show()
```
